```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
results_df = pd.read_csv('ensemble_results_df.csv')
results_df_2 = pd.read_csv('ensemble_results_df_2.csv')
results = pd.concat([results_df, results_df_2])
results["ensemble_name"] = ["hoeffdingtreeclassifier", "extremelyfastdecisiontreeclassifier"] * 10
results["model_name"] = ['BaggingClassifier', 'AdaBoostClassifier', 'ADWINBaggingClassifier', 'SRPClassifier', 'LeveragingBaggingClassifier'] * 4
results = results.reset_index(drop=True)
```

```python
results
```

```python
model_accuracy = results.groupby('model_name')['accuracy'].mean()
fig = plt.figure(figsize=(12, 6))
sns.barplot(x=model_accuracy.index, y=model_accuracy.values)
plt.title('Average accuracy by ensemble')
plt.xlabel('Ensemble name')
plt.ylabel('Average accuracy')
plt.ylim([0.4, 0.6])
plt.savefig("ensemble_analysis_images/accuracy_by_ensemble.png")
```

```python
model_accuracy = results.groupby('ensemble_name')['accuracy'].mean()
fig = plt.figure(figsize=(12, 6))
sns.barplot(x=model_accuracy.index, y=model_accuracy.values)
plt.title('Average accuracy by model')
plt.xlabel('Model name')
plt.ylabel('Average accuracy')
plt.ylim([0.4, 0.6])
plt.savefig("ensemble_analysis_images/accuracy_by_base_model_count.png")
```

```python
results["model_count"] = [5] * 10 + [10] * 10
```

```python
model_accuracy = results.groupby('model_count')['accuracy'].mean()
fig = plt.figure(figsize=(12, 6))
sns.barplot(x=model_accuracy.index, y=model_accuracy.values)
plt.title('Average accuracy by base model count')
plt.xlabel('Base model count')
plt.ylabel('Average accuracy')
plt.ylim([0.4, 0.6])
plt.savefig("ensemble_analysis_images/accuracy_by_ensemble.png")
```
